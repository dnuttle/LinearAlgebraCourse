# VIDEO: Vector-vector multiplication: the dot product
---


```python
## many ways to compute the dot product

v1 = np.array([ 1, 2, 3, 4, 5, 6 ])
v2 = np.array([ 0, -4, -3, 6, 5, 9 ])

# method 1
dp = sum( np.multiply(v1,v2) )

# method 2
dp = np.dot( v1,v2 )

# method 3
dp = np.matmul( v1,v2 )

# method 4
dp = 0  # initialize

# loop over elements
for i in range(0,len(v1)):
    
    # multiply corresponding element and sum
    dp = dp + v1[i]*v2[i]
    
print(dp)
```

# VIDEO: The dot product from a geometric perspective
---


```python

# two vectors
v1 = np.array([ 2,  4, -3 ])
v2 = np.array([ 0, -3, -3 ])

# compute the angle (radians) between two vectors
ang = np.arccos( np.dot(v1,v2) / (np.linalg.norm(v1)*np.linalg.norm(v2)) )


# draw them
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot([0, v1[0]],[0, v1[1]],[0, v1[2]],'b')
ax.plot([0, v2[0]],[0, v2[1]],[0, v2[2]],'r')

plt.axis((-6, 6, -6, 6))
plt.title('Angle between vectors: %s rad.' %ang)
plt.show()
```

```python
## equivalence of algebraic and geometric dot product formulas

# two vectors
v1 = np.array([ 2,  4, -3 ])
v2 = np.array([ 0, -3, -3 ])


# algebraic
dp_a = np.dot( v1,v2 )

# geometric
dp_g = np.linalg.norm(v1)*np.linalg.norm(v2)*np.cos(ang)

# print dot product to command
print(dp_a)
print(dp_g)
```

```python
#Quiz 3: Relative vector angles
def angbetween(x, y):
    return np.arccos(np.dot(x, y) / (np.linalg.norm(x) * np.linalg.norm(y)))
a = np.array([1,-2])
b = np.array([2,3])
c = np.array([0,2])
print("a and b")
print(np.dot(a, b))
print("dot product < 0, obtuse")
print(angbetween(a, b))
print("angle > pi/2, obtuse")
print("a and c")
print(np.dot(a, c))
print("dot product < 0, obtuse")
print(angbetween(a, c))
print("angle > pi/2, obtuse")
print("b and c")
print(np.dot(b, c))
print("dot product > 0, acute")
print(angbetween(b, c))
print("angle < pi/2, acute")
```

### Coding challenge
Test whether the dot product sign is invariant to scalar multiplication. First problem is that it is not immediately obvious what he means. It sounds like he wants to know if multiplying 1 or -1 with the result is the same.

Actually the point was that the sign of the product of the two scalars will be applied to the dot product, so that if the scalar product sign is negative, then the sign of the dot product is reversed. The one exception is when the dot product is zero, then the scalars have no effect.

Geometric explanation: If the dot product is negative, it means the angle between the vectors is obtuse. If one of the scalars is negative, it reverses the direction of that vector, and the angle is now acute. If both scalars are negative, it reverses both vectors, so the angle is unchanged. In any case, the length of the vectors does not affect the angle between them.

A key point to remember for future reference: If two vectors are orthogonal, the dot product is zero, and therefore scalar multiplication has no effect at all on them.

```python

v1 = np.array([  1, 2, 3 ])
v2 = np.array([ -1, 0, 1 ])

# outer product
print(np.outer(v1,v2))

# terrible programming, but helps conceptually:
op = np.zeros((len(v1),len(v1)))
for i in range(0,len(v1)):
    for j in range(0,len(v2)):
        op[i,j] = v1[i] * v2[j]

print(op)
```
